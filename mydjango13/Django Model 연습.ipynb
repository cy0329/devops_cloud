## Get 1 Row

```python
pk=10
not_existed_pk = 100000
```

```python
shop = Shop.objects.get(pk=pk)  # 가져올 대상이 1개임을 기대. 1개가 아니면 예외가 발생.
shop
```

```python
Shop.objects.get(pk=not_existed_pk)  # Shop.DoesNotExist 예외가 발생
```

```python
Shop.objects.filter(pk=pk)  # 가져올 대상이 여러개임을 기대
```

```python
Shop.objects.filter(pk=pk)[0]
shop
```

```python
Shop.objects.filter(pk=not_existed_pk)[0]  # IndexError 예외가 발생( [0] 번 인덱스로 접근했기 때문)
```

```python
Shop.objects.filter(pk=not_existed_pk).first()  # 지정 row가 없다면 None을 반환할 뿐, 예외는 발생X.
```

## Conditions (OR)

```python
qs = Shop.objects.all()
qs.filter(pk__gte=5, pk__lt=10)  # AND
```

```python
from django.db.models import Q
```

```python
Q(pk__gte=5)
```

```python
Q(pk__lt=10)
```

```python
Q(pk__gte=5) & Q(pk__lt=10)
```

```python
Q(pk__gte=5) | Q(pk__lt=10)
```

```python
qs = Shop.objects.all()
qs.filter(
    Q(pk__gte=5) | Q(pk__lt=10)
)
```

```python
conditions = Q(pk__gte=5)
conditions = conditions | Q(pk__lt=10)

conditions
```

## 지정 범위의 Rows

```python
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
qs.update(
    description="not used description",
)
```

```python
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
for shop in qs:
    shop.description = "not used description"
    shop.save()
```
